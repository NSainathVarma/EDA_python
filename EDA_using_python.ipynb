```python
new = df.drop(['Generation', 'Legendary'],axis = 1)
new.head()
```

```python
#Relationship analysis
##correlation matrix

correlation = new.corr()

sns.heatmap(correlation, xticklabels = correlation.columns, yticklabels = correlation.columns,annot = True)

```

```python
#pair plotting
sns.pairplot(new) 
```

```python
#scatterplot
sns.relplot(x = 'Attack', y = 'Defense', hue = 'Sp. Atk', data = new)
```

```python
#Histogram
sns.displot(new['Attack'],bins = 15)#bins specifies the number of bars in the graph
```

```python
#Box plot

sns.catplot(x = 'HP', kind = 'box', data = new)
```
